```python
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import preprocessing
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
```

```python
def func(trainer, train_x,train_y,test_x,test_y):
    print("-"*15,end=' ')
    print(trainer.__class__.__name__,end=' ')
    print("-"*15)
    print("-"*54)
    trainer.fit(train_x,train_y)
    pred_y = trainer.predict(test_x)
    print(classification_report(test_y,pred_y))
    print("-"*54)
```

```python
def func2(loaded_data):
    data = loaded_data.data
    label = loaded_data.target

    loaded_data.target_names

    df = pd.DataFrame(
        data=data,
        columns = loaded_data.feature_names
    )
    df["label"]=label
    print("*"*40, "data", end=" ")
    print("*"*40,"\n")
    print(df)
        
    X_train,X_test,Y_train,Y_test = train_test_split(

        data,
        label,
        test_size=0.2,
        random_state = 7
    )

    X_train = preprocessing.scale(X_train)
    X_test = preprocessing.scale(X_test)

    func(DecisionTreeClassifier(random_state=32),
         X_train,
         Y_train,
         X_test,
         Y_test
        )
    func(RandomForestClassifier(random_state=32),
         X_train,
         Y_train,
         X_test,
         Y_test
        )
    func(svm.SVC(),
         X_train,
         Y_train,
         X_test,
         Y_test
        )
    func(SGDClassifier(random_state=32),
         X_train,
         Y_train,
         X_test,
         Y_test
        )
    func(LogisticRegression(random_state=32),
         X_train,
         Y_train,
         X_test,
         Y_test
        )
    print("\n\n")
```

```python
func2(load_digits())
func2(load_wine())
func2(load_breast_cancer())
```

load_data: 테스트 데이터가 클래스별로 균형있게 존재하므로 accuracy를 평가지표로 사용한다.(특별히 중요한 클래스가 없으므로 recall은 패스)
load_wine: 테스트 데이터가 클래스별로 불균형하게 존재하므로 precision을 평가지표로 사용한다. (특별히 중요한 클래스가 없으므로 recall은 패스)
load_breast_cancer: 양성의 경우를 확실하게 검출해야 하므로 recall을 평가 지표로 사용한다.(암의 존재에 대한 이진 분류(true/false)에서 특히, true에 대한 정확도가 중요하므로 recall이 중요한 것이다.)
